```python
import torch
import random
import os
import torchvision.models as models

os.environ["CUDA_VISIBLE_DEVICES"] = '1'

path = '/home/bengio/NAS_AIlab_dataset/personal/heo_yunjae/uncertainty-activelearning/SSAL/SimCLR-2/runs/Sep18_23-41-06_ailab-server-bengio/checkpoints/model.pth'
state_dict = torch.load(path)
model = models.resnet18()
print(state_dict.keys())
new_stdict = dict()
for key1 in state_dict.keys():
# key1 = 'features.4.0.conv1.weight'
    key1_list = key1.split('.')
    if(key1_list[0]=='features'):
        if len(key1_list) == 3:
            if key1_list[1] == '0':
                key2 = 'conv1.weight'
            if key1_list[1] == '1':
                key2 = 'bn1.'+key1_list[-1]
        elif len(key1_list) >= 5:
            key2 = 'layer'+f'{int(key1_list[1])-3}.'+'.'.join(key1_list[2:])
    new_stdict[key2] = state_dict[key1]
print(new_stdict.keys())
# print(state_dict['features.4.0.conv1.weight'].shape)
model.load_state_dict(new_stdict, strict=False)
```
